```python
from pathlib import Path
import shutil
from tkinter import Tk, filedialog

# =========================================
# FIND PROJECT ROOT
# =========================================

project_root = Path.cwd().parent
keys_folder = project_root / "Keys"
keys_folder.mkdir(exist_ok=True)

# =========================================
# SELECT BIGQUERY JSON FILE
# =========================================

root = Tk()
root.withdraw()

print("Please select your BigQuery JSON key file...")

json_file = filedialog.askopenfilename(
    title="Select BigQuery Service Account JSON",
    filetypes=[("JSON Files", "*.json")]
)

if not json_file:
    raise Exception("No JSON file selected.")

destination_json = keys_folder / "BigQueryKey.json"

# Overwrite existing file
if destination_json.exists():
    destination_json.unlink()

shutil.copy2(json_file, destination_json)

print(f"✓ BigQuery key saved to:\n{destination_json}")

# =========================================
# ENTER POSTGRES URL
# =========================================

postgres_url = input(
    "\nPaste your PostgreSQL connection URL and press Enter:\n"
).strip()

if not postgres_url:
    raise Exception("PostgreSQL URL cannot be empty.")

# =========================================
# CREATE/OVERWRITE PostGresKey.py
# =========================================

postgres_file = keys_folder / "PostGresKey.py"

postgres_file.write_text(
    f'POSTGRES_URL = "{postgres_url}"\n',
    encoding="utf-8"
)

print(f"\n✓ PostgreSQL configuration saved to:\n{postgres_file}")

print("\nSetup completed successfully!")
```
