```python
from pathlib import Path
import shutil
import tkinter as tk
from tkinter import filedialog, messagebox

# =========================================
# FIND PROJECT ROOT
# =========================================

project_root = Path.cwd().parent
keys_folder = project_root / "Keys"
keys_folder.mkdir(exist_ok=True)

# =========================================
# SELECT BIGQUERY JSON FILE
# =========================================

root = tk.Tk()
root.withdraw()
root.wm_attributes("-topmost", 1)
root.update()

json_file = filedialog.askopenfilename(
    parent=root,
    title="Select BigQuery Service Account JSON",
    filetypes=[("JSON Files", "*.json")]
)

root.destroy()

if not json_file:
    raise Exception("No JSON file selected.")

destination_json = keys_folder / "BigQueryKey.json"

# Remove old file if exists
if destination_json.exists():
    destination_json.unlink()

# Copy selected file
shutil.copy2(json_file, destination_json)

print(f"✓ BigQuery key saved to:\n{destination_json}")

# =========================================
# POSTGRES URL POPUP WINDOW
# =========================================

postgres_result = {"url": None}

app = tk.Tk()
app.title("Database Setup")
app.geometry("800x150")
app.resizable(False, False)
app.attributes("-topmost", True)

label = tk.Label(
    app,
    text="Paste PostgreSQL Connection URL:",
    font=("Arial", 10)
)
label.pack(pady=(15, 5))

entry = tk.Entry(app, width=120)
entry.pack(padx=15)

entry.focus_set()

def save_url():
    postgres_result["url"] = entry.get().strip()

    if not postgres_result["url"]:
        messagebox.showerror(
            "Missing Value",
            "PostgreSQL URL cannot be empty."
        )
        return

    app.destroy()

save_btn = tk.Button(
    app,
    text="Save",
    width=15,
    command=save_url
)
save_btn.pack(pady=15)

app.bind("<Return>", lambda event: save_url())

app.mainloop()

postgres_url = postgres_result["url"]

if not postgres_url:
    raise Exception("PostgreSQL URL cannot be empty.")

# =========================================
# CREATE/OVERWRITE PostGresKey.py
# =========================================

postgres_file = keys_folder / "PostGresKey.py"

# Remove old file if exists
if postgres_file.exists():
    postgres_file.unlink()

postgres_file.write_text(
    f'POSTGRES_URL = "{postgres_url}"\n',
    encoding="utf-8"
)

print(f"✓ PostgreSQL configuration saved to:\n{postgres_file}")

# =========================================
# SUCCESS MESSAGE
# =========================================

success_root = tk.Tk()
success_root.withdraw()
success_root.attributes("-topmost", True)

messagebox.showinfo(
    "Setup Complete",
    "BigQuery and PostgreSQL credentials have been configured successfully."
)

success_root.destroy()

print("\nSetup completed successfully!")
```
